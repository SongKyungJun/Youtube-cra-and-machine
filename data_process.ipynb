```python
def bin_and_mean_index(a_df: pd.DataFrame, bin_amount: int):
    bins = pd.cut(a_df['subscribers'], bins=bin_amount, labels=range(bin_amount)).rename("subscriber_bin")
    groupped = pd.concat([a_df, bins], axis=1).groupby(by='subscriber_bin')
    mean_v_index = groupped.count()
    mean_v_sum = mean_v_index['index'].cumsum()
    return mean_v_sum

def bin_and_mean_count(a_df: pd.DataFrame, bin_amount: int):
    bins = pd.cut(a_df['subscribers'], bins=bin_amount, labels=range(bin_amount)).rename("subscriber_bin")
    groupped = pd.concat([a_df, bins], axis=1).groupby(by='subscriber_bin')
    mean_v_count = groupped.mean()
    return mean_v_count
```
